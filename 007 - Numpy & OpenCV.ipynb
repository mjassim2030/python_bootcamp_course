## Warp Perspective

```python
# Read the image and change BGR to RGB
img = cv2.imread("media/card.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# we need 4 points and based on these 4 points we will apply wrap perspective
width, height = 250, 350
```

```python
# Define out 4 corner points of the card (you can get them from the paint application)
points1 = np.float32([[361,64],[473,16],[582,147],[459,199]])
points2 = np.float32([[0,0],[width,0],[width,height],[0,height]])
```

```python
# Transformation matrix which is required by the perpective itself
matrix = cv2.getPerspectiveTransform(points1, points2)
imgOutput = cv2.warpPerspective(img, matrix, (width, height))

plt.imshow(imgOutput)
```

## Face Detection

```python
# Read and show image file
img = cv2.imread("media/man.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(img)
```

```python
faceCascade = cv2.CascadeClassifier("media/haarcascade_frontalface_default.xml")
```

```python
imgGray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
faces = faceCascade.detectMultiScale(imgGray, 1.15, 2)
```

```python
# Draw square on all faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
```

```python
plt.imshow(img)
```

## Eye Detection

```python
Cascade = cv2.CascadeClassifier("media/haarcascade_eye.xml")

# Read and show image file
img = cv2.imread("media/man.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

imgGray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
objects = Cascade.detectMultiScale(imgGray, 1.15, 2)

# Draw square on all faces
for (x, y, w, h) in objects:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

plt.imshow(img)
```

## Car Plate Number Detection

```python
# Read and show image file
img = cv2.imread("media/car.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
plateCascade = cv2.CascadeClassifier("media/haarcascade_russian_plate_number.xml")
imgGray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
numberPlates = plateCascade.detectMultiScale(imgGray, 1.1, 10)
```

```python
# Draw square on all plate numbers
for (x, y, w, h) in numberPlates:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
```

```python
plt.imshow(img)
```

## People Detection in a Video

```python
import PIL.Image
from io import BytesIO
import IPython.display
from IPython.display import clear_output
```

```python
Cascade = cv2.CascadeClassifier("media/haarcascade_fullbody.xml")
```

```python
cam = cv2.VideoCapture("media/garden.avi")
```

```python
while(True):
    # Capture frame-by-frame
    error, frame = cam.read()

    # Convert the image from OpenCV BGR format to matplotlib RGB format
    imgGray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    people = Cascade.detectMultiScale(imgGray, 1.075, 3)

    for (x, y, w, h) in people:
        if  3500 < w * h < 7000:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
            cv2.putText(frame,"People Counter: " + str(len(people)),(0,frame.shape[0] - 10),cv2.FONT_HERSHEY_COMPLEX_SMALL,1,(0,255,0),1)
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    f = BytesIO()
    PIL.Image.fromarray(frame).save(f, "jpeg")
    IPython.display.display(IPython.display.Image(data=f.getvalue()))

    # Display the frame until new frame is available
    clear_output(wait=True)
```
